## Configure Gemini API

This section initializes the Gemini model for generating AI-powered interview questions.

```python
import google.generativeai as genai

# Paste your Gemini API key below
API_KEY = "Enter your gemini API key"

genai.configure(api_key=API_KEY)

model = genai.GenerativeModel("gemini-2.5-flash")

print("✅ Gemini API configured successfully!")
```

```python
def generate_interview_questions(role, job_description, difficulty="Medium"):
    prompt = f"""
You are an expert technical interviewer.

Generate interview questions for the following internship role.

Role:
{role}

Job Description:
{job_description}

Difficulty:
{difficulty}

Generate:

1. 5 Technical Questions
2. 5 Behavioral Questions

Return the output with proper headings.
"""

    response = model.generate_content(prompt)
    return response.text

print("✅ Interview Question Generator is ready!")
```

```python
role = "Machine Learning Intern"

job_description = """
Looking for an intern with knowledge of Python, Machine Learning,
Data Analysis, Pandas, NumPy and Scikit-learn.
"""

result = generate_interview_questions(role, job_description)

print(result)
```

## User Input

The user provides the internship role, job description, and difficulty level. These inputs are used by the AI model to generate customized interview questions.

```python
# User Inputs

role = input("Enter Internship Role: ")

difficulty = input("Enter Difficulty (Easy/Medium/Hard): ")

print("\nEnter the Job Description.")
print("When you're finished, press Enter twice.\n")

lines = []
while True:
    line = input()
    if line == "":
        break
    lines.append(line)

job_description = "\n".join(lines)
```

```python
questions = generate_interview_questions(
    role,
    job_description,
    difficulty
)

print(questions)
```

## Generate Interview Questions

This function generates customized interview questions using the Gemini model and includes basic error handling.

```python
def generate_interview_questions(role, job_description, difficulty="Medium"):
    """
    Generate interview questions using Gemini AI.

    Parameters:
        role (str): Internship role
        job_description (str): Job description
        difficulty (str): Easy, Medium, or Hard

    Returns:
        str: Generated interview questions
    """

    prompt = f"""
You are an experienced technical interviewer.

Generate interview questions for the following internship role.

Role:
{role}

Job Description:
{job_description}

Difficulty:
{difficulty}

Generate:
1. Five Technical Interview Questions
2. Five Behavioral Interview Questions

Format the output clearly using headings and numbering.
"""

    try:
        response = model.generate_content(prompt)
        return response.text

    except Exception as e:
        return f"Error: {e}"
```
